### Loss function every 10 iteration 

```python
import pandas as pd
import matplotlib.pyplot as plt

df_i = pd.read_csv("iter_losses_every_10_no_process.csv")

max_iter_per_epoch = df_i['iteration'].max()
df_i['global_iter'] = (df_i['epoch'] - 1) * max_iter_per_epoch + df_i['iteration']

df_i = df_i.sort_values('global_iter')

plt.figure(figsize=(20, 5))
plt.plot(df_i['global_iter'], df_i['loss'], marker='o')
plt.xlabel('Iteration (global)')
plt.ylabel('Loss')
plt.title('Loss vs Iteration')
plt.grid(True)
plt.show()
```

### Loss every epoch

```python
df_e = pd.read_csv('epoch_metrics_no_process.csv')

plt.figure(figsize=(8, 5))

plt.plot(df_e['epoch'], df_e['train_loss'], marker='o', label='Train Loss')
plt.plot(df_e['epoch'], df_e['val_loss'], marker='o', label='Validation Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train vs Validation Loss')
plt.legend()
plt.grid(True)

plt.show()
```
